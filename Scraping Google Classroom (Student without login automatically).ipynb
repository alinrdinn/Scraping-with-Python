```python
# Import Selenium and Time
from selenium import webdriver
import time
```

```python
# Open Web Driver (Chrome)
Driver = webdriver.Chrome()
Driver.get("https://classroom.google.com/u/0/h")
time.sleep(10)
```

**Enter email and password manually.**

```python
def goback(how_many_times):
    for i in range(0, how_many_times):
        Driver.back()
        time.sleep(10)
    
def assignment_tab():
    tabs = Driver.find_elements_by_css_selector("div.wZTANe")
    return (tabs[1].find_element_by_tag_name("a")).get_attribute("href")

def member_tab():
    tabs = Driver.find_elements_by_css_selector("div.wZTANe")
    return (tabs[2].find_element_by_tag_name("a")).get_attribute("href")

def scroll(n):
    for i in range(0, n):
        time.sleep(4)
        Driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        try:
            Button = Driver.find_element_by_css_selector("div.uArJ5e.UQuaGc.kCyAyd.nZ34k")
            Button.click()
        except:
            break
        time.sleep(4)

def scraping_assignment():
    list_of_assignments = []
    
    Assignments = Driver.find_elements_by_class_name("jHgkif.raZwr")
    
    for Assignment in Assignments:
        Assignments1 = Assignment.find_elements_by_css_selector('li[guidedhelpid="classworkTopicListGh"]')
        for Assign in Assignments1:
            assignment_name = Assign.find_element_by_css_selector("span.YVvGBb.UzbjTd").text
            date = Assign.find_element_by_css_selector("div.wCDkmf.T3FW5d.YVvGBb.dDKhVc").text.split("Tenggat: ")
            list_of_assignments.append([assignment_name, date[len(date)-1]])

    return list_of_assignments

def scraping_member():
    teachers = Driver.find_element_by_css_selector("div.pEwOBc.FvI3qc.raZwr").text.split("\n")
    students = Driver.find_element_by_css_selector("div.pEwOBc.HTxhwc").text.split("\n") 
    
    return {
        'Teachers' : teachers[1:],
        'Students' : students[2:]
    }
    
def scraping_subject():
    
    # Subject's info (1)
    subject_name = Driver.find_element_by_id("UGb2Qe").text
    
    # Goto Assignment tab
    tab = assignment_tab()
    Driver.get(tab)
    
    time.sleep(10)
    scroll(6) # Scroll 6 times in the Assignment tab
    
    # Subject's Info (2)
    assignments = scraping_assignment()
    
    # Goto Member tab
    tab = member_tab()
    Driver.get(tab)
    
    time.sleep(10)
    scroll(6) # Scroll 6 times in the Member tab
    
    #Subject's Info (3)
    member = scraping_member()
    
    return {
        'subject name' : subject_name,
        'assignments' : assignments,
        'members' : member
    }
```

```python
number_of_classes = len(Driver.find_elements_by_css_selector("div.Tc9hUd"))
list_of_subjects = []
for i in range(0, number_of_classes):
    # Entering Class/Subject
    classes = Driver.find_elements_by_css_selector("div.Tc9hUd")
    classes[i].click()
    
    # Scraping Subject
    time.sleep(5)
    list_of_subjects.append(scraping_subject())
    
    goback(3)
```
